```python
from sklearn import datasets

datos = datasets.load_breast_cancer(as_frame=True) #Nos permite cargar los datos en la variable data directamente como un dataframe de pandas
```

```python
#Creemos un modelo para predecir
#-------------------------------
#Separemos los datos
from sklearn.model_selection import train_test_split
#En este data set no debemos separar la columna en cuestiÃ³n, ya viene separada por defecto
X = datos.data
y = datos.target
X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size = 0.5)
```

```python
#Creemos nuestro modelo usando el support vector classifier SCV
from sklearn import svm
from sklearn import metrics

#Evaluemos los hiper-parÃ¡metros
kernels = ['linear', 'rbf', 'sigmoid']
gammas = [1, .1, .01, .001, .0001, .00001]

for kernel in kernels:
    for gamma in gammas:
        modelo = svm.SVC(kernel = kernel, gamma = gamma)
        modelo.fit(X_ent, y_ent)
        predicciones = modelo.predict(X_pru)
        print(f"Kernel: {kernel}; Gamma: {gamma}; Exactitud: ", metrics.accuracy_score(y_pru, predicciones))
        print(f"Kernel: {kernel}; Gamma: {gamma}; PrecisiÃ³n: ", metrics.precision_score(y_pru, predicciones))
```

```python
#De lo anterior vemos que para este set de datos especÃ­fico el kernel ideal parece ser el lineal y nos quedamos con este kernel
modelo = svm.SVC(kernel = "linear") #Como a este kernel no le afecta el valor de gamma, lo dejamos asÃ­
modelo.fit(X_ent, y_ent)
predicciones = modelo.predict(X_pru)
print(f"Kernel: {kernel}; Gamma: {gamma}; Exactitud: ", metrics.accuracy_score(y_pru, predicciones))
print(f"Kernel: {kernel}; Gamma: {gamma}; PrecisiÃ³n: ", metrics.precision_score(y_pru, predicciones))
```

```python
#Hagamos ahora el reporte de clasificaciÃ³n
from sklearn.metrics import classification_report
print(classification_report(y_pru, predicciones))
```

```python
#Imprimimos la matriz de consusiÃ³n
from sklearn.metrics import confusion_matrix
import pandas as pd
pd.DataFrame(confusion_matrix(y_pru, predicciones))
```
